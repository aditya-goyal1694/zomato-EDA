```python
country_df=pd.read_excel("C:/Users/adity/Downloads/Zomatodataset/Country-Code.xlsx")
country_df
```

```python
df=pd.merge(zomato_df, country_df, on="Country Code", how="left")
df
```

```python
df.columns
```

```python
df.Country.value_counts()
```

```python
country_names=df.Country.value_counts().index
country_names
```

```python
country_val=df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val, labels=country_names, autopct='%.2f%%')
```

```python
#Top 3
plt.pie(country_val[:3], labels=country_names[:3], autopct='%.2f%%')
```

#### Observation
- More than 90% of zomato sales is from India. Followed by 4.5% from US and 0.8% from UK. Rest all countries contributes less than 1% to the sales.

```python
ratings=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue='Rating text', palette=['grey', 'red', 'orange', 'yellow', 'green', 'olive'])
```

#### Observation:
- Not rated count is very high.
- Most of the ratings are between 2.5 to 4.4
